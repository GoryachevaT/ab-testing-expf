Формула линеаризации

$$
L x,y,k(U) = X(u) - kY(u)
$$

ГДЕ; x-сумма действий, Y - кол-во действий, K - общее RATIO по контролю

```python
def linearization (control,experiment):
for row in control:
numerator +=sum(row)
denominator +=len(row)

control_mean=numerator/denominator
new_control=[sum(row)-len(row)*control_mean in control]
experiment=[sum(row)-len(row)*control_mean in experiment]
return new_control, experiment
```
